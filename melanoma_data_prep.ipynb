```python
import pandas as pd
train_csv_path = 'train.csv'
data_dir = 'C:/Users/SAANKHYA/Downloads/archive/train'
train_df = pd.read_csv(train_csv_path)
train_df.head()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

row = train_df.iloc[45]
print(row)
img = Image.open(f"{data_dir}/{row[0]}.png")
plt.imshow(img); plt.show()
```

```python
import os
folderlocation = 'Melanoma/'

if not os.path.exists(folderlocation):
    os.mkdir(folderlocation)

path = folderlocation+'train/'
if not os.path.exists(path):
    os.mkdir(path)
path = folderlocation+'train/benign'
if not os.path.exists(path):
    os.mkdir(path)
path = folderlocation+'train/malignant'
if not os.path.exists(path):
    os.mkdir(path)
```

```python
from shutil import copyfile
for row in train_df.iloc:
    src = f"{data_dir}/{row[0]}.png"
    des = folderlocation+'train/'+row[6]+'/'+row[0]+'.png'
    copyfile(src, des)
```
